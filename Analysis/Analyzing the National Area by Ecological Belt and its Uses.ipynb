## Analyzing the National Area by Ecological Belt and its Uses

Total Area of Nepal - 1,47,181 sq.km.

### Analyzing the National Area by Ecological Belt

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
load_data = pd.read_csv('Data/National Area by Ecological Belt.csv')
```

```python
load_data
```

```python
label= load_data['BELT']
```

```python
values = load_data['AREA(SQUARE KM)']
```

```python
explode = [0.03,0.01,0]
```

```python
colors = ['pink','cyan','tan']
```

```python
fig = plt.figure(figsize = (7,7))
plt.pie(values,
        labels = label,
        explode = explode,
        shadow=False,
        colors = colors,
        autopct='%1.1f%%',
        textprops={'fontsize': 14});
plt.title('\n\n\n\nNational Area by Ecological Belt of Nepal', 
          color='Purple',
         fontsize=18);
```

### Land Use Distribution by Use Category

```python
land_use = pd.read_csv('Data/Land Use Distribution by Use Category.csv')
```

```python
land_use
```

```python
total_land_use = sum(land_use['Area (Hectares)'])
```

```python
print('Total area of Nepal in Hectares is: ', total_land_use)
```

```python
category_label = land_use['Category']
area_values = land_use['Area (Hectares)']
col = ['tan','violet','silver','orange','teal','purple','pink']
explodes = [0.0,0,0,0.0,0,0,0]
fig = plt.figure(figsize=[10,10])
plt.pie(area_values, 
        explode = explodes,
        colors=col,
        labels = category_label,
        autopct='%1.01f%%',
        textprops={'fontsize': 12});
plt.title('\nLand Use distribution by Used Category of Nepal',fontsize=18, color='purple');
```
